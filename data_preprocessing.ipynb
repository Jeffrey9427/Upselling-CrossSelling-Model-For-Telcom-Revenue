```python
concatenated_df = pd.concat([df, df1], ignore_index=True)

concatenated_df
```

```python
print(type(concatenated_df))
```

```python
total_rows = concatenated_df.shape[0]
print(f'Total rows in the DataFrame: {total_rows}')
```

```python
for column in concatenated_df.columns:
    count = concatenated_df[column].count()
    print(f"Column '{column}' has {count} data points.")
```

```python
for column in concatenated_df.columns:
    missing_count = df[column].isna().sum()
    total_count = len(df)
    missing_percentage = (missing_count / total_count) * 100
    print(f"Column '{column}' has {missing_percentage:.2f}% missing data points.")
```

```python
for column in concatenated_df.columns:
    concatenated_df[column] = concatenated_df[column].fillna("not found")

concatenated_df
```

```python
# Convert 'id' to integer and replace non-integer values with 'NULL'
concatenated_df['id'] = pd.to_numeric(concatenated_df['id'], errors='coerce').fillna('null').astype(int)
concatenated_df
```

```python
concatenated_df['created_at'] = pd.to_datetime(df['created_at'])
concatenated_df['created_at'] = concatenated_df['created_at'].dt.strftime('%Y-%m-%d %H:%M:%S')
concatenated_df['created_at'] = concatenated_df['created_at'].fillna('not found')

concatenated_df
```

```python
#revenue
concatenated_df['revenue'] = concatenated_df['revenue'].apply(lambda x: 0 if isinstance(x,str) and ('-' in x or ':' in x) else float(x))
concatenated_df
```

```python
# Currency
concatenated_df['currency'] = concatenated_df['currency'].str.lower()
concatenated_df
```

```python
# Operator
concatenated_df['operator'] = concatenated_df['operator'].str.lower()
concatenated_df
```

```python
#service_price
concatenated_df['service_price'] = concatenated_df['service_price'].apply(lambda x: 0 if isinstance(x,str) and ('-' in x or ':' in x) else float(x))
concatenated_df
```

```python

concatenated_df['subs_date'] = pd.to_datetime(concatenated_df['subs_date'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')

# Format the datetime column without timezone information
concatenated_df['subs_date'] = concatenated_df['subs_date'].dt.strftime('%Y-%m-%d %H:%M:%S')

print(concatenated_df['subs_date'].head(100))
```

```python
concatenated_df['subs_date'] = pd.to_datetime(concatenated_df['subs_date'], errors='coerce')

# Find rows with '9999-01-01' in the time portion
rows_with_9999_time = concatenated_df[concatenated_df['subs_date'].dt.strftime('%H:%M:%S') == '9999-01-01']

# Display the rows with the specified time
print(rows_with_9999_time)
```

```python
concatenated_df['renewal_date'] = pd.to_datetime(concatenated_df['renewal_date'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')

# Format the datetime column without timezone information
concatenated_df['renewal_date'] = concatenated_df['renewal_date'].dt.strftime('%Y-%m-%d %H:%M:%S')

print(concatenated_df['renewal_date'].head(100))
```

```python

concatenated_df['freemium_end_date'] = pd.to_datetime(concatenated_df['freemium_end_date'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')

# Format the datetime column without timezone information
concatenated_df['freemium_end_date'] = concatenated_df['freemium_end_date'].dt.strftime('%Y-%m-%d %H:%M:%S')

print(concatenated_df['freemium_end_date'].head(100))
```

```python
# profile_status
concatenated_df['profile_status'] = concatenated_df['profile_status'].str.lower()

# Define a list of valid values
valid_unsubs_values = ['active', 'grace', 'parking', 'purging', 'blacklist']

# Replace values that are not in the valid list with 'null'
concatenated_df['profile_status'] = concatenated_df['profile_status'].apply(lambda x: x if x in valid_unsubs_values else 'not found')

concatenated_df
```

```python
# Source
concatenated_df['source'] = concatenated_df['source'].str.lower()
concatenated_df
```

```python

unique_countries = concatenated_df['country'].unique()

# Display the unique values
print("Unique values in the 'country' column:")
print(unique_countries)
```

```python
# Create a mapping of country codes to country names
country_mapping = {
    'TH': 'thailand',
    'OM' : 'oman',
    'LA' : 'laos',
    'SN' : 'senegal',
    'ID' : 'indonesia',
    'PH' : 'philippines'
}

# Replace the country codes with country names
concatenated_df['country'] = concatenated_df['country'].replace(country_mapping)
concatenated_df
```

```python
# Assuming concatenated_df is your DataFrame
unique_cycle = concatenated_df['cycle'].unique()

# Display the unique values
print("Unique values in the 'cycle' column:")
print(unique_cycle)
```
